```
import pandas as pd
import numpy as np

DATA_DIR = "../../data/raw/"
CAST_FILE = DATA_DIR + "Cast.csv"
BOTTLE_FILE = DATA_DIR + "Bottle.csv"

# Load Cast (small, load fully)
cast = pd.read_csv(CAST_FILE, low_memory=False)
print(f"Cast shape: {cast.shape}")
cast.head(3)
```

```
# Check key Cast columns
key_cast_cols = ['Cst_Cnt', 'Sta_ID', 'Year', 'Month', 'Lat_Dec', 'Lon_Dec']
print("Cast key columns present:", [c for c in key_cast_cols if c in cast.columns])
print("\nMissing key cast cols:", [c for c in key_cast_cols if c not in cast.columns])
print("\nAll Cast columns:")
print(cast.columns.tolist())
```

```
# Load Bottle - just first few rows + columns first to understand structure
bottle_sample = pd.read_csv(BOTTLE_FILE, nrows=5, low_memory=False)
print(f"Bottle columns ({len(bottle_sample.columns)}):")
print(bottle_sample.columns.tolist())
```

```
bottle_sample = pd.read_csv(BOTTLE_FILE, nrows=5, low_memory=False, encoding='latin-1')
print(f"Bottle columns ({len(bottle_sample.columns)}):")
print(bottle_sample.columns.tolist())
```

```
# Full bottle load
print("Loading full Bottle CSV (~175MB)...")
bottle = pd.read_csv(BOTTLE_FILE, low_memory=False, encoding='latin-1')
print(f"Bottle shape: {bottle.shape}")
bottle.head(3)
```

```
# Quality code columns for our 5 variables
# T_qual, O_qual, (pH has no qual col?), NO3q, PO4q
qual_cols = ['T_qual', 'O_qual', 'NO3q', 'PO4q']
print("Quality code columns value counts:")
for col in qual_cols:
    print(f"\n{col}:")
    print(bottle[col].value_counts().head(10))
```

```
# pH has no quality column — check pH1 nulls directly
# Also check if there's a Chlqua for ChlorA
print("pH1 null count:", bottle['pH1'].isna().sum(), "/ total:", len(bottle))
print("Chlqua value counts:")
print(bottle['Chlqua'].value_counts().head())

# Also check: are quality codes strings or floats?
print("\nT_qual dtype:", bottle['T_qual'].dtype)
print("Sample T_qual values:", bottle['T_qual'].dropna().unique()[:10])
```

```
# pH1 is almost entirely null (895287/895371). Check pH2 as well.
print("pH2 null count:", bottle['pH2'].isna().sum())
print("\npH1 non-null sample:")
print(bottle[bottle['pH1'].notna()][['Cst_Cnt','pH1','pH2']].head(10))
```

```
# Coverage summary for all 5 key variables
vars_of_interest = ['T_degC', 'O2ml_L', 'pH1', 'NO3uM', 'PO4uM', 'ChlorA', 'SiO3uM']
print("Variable coverage (non-null rows):")
for v in vars_of_interest:
    n = bottle[v].notna().sum()
    pct = 100 * n / len(bottle)
    print(f"  {v:<12} {n:>8,}  ({pct:.1f}%)")
```

```
# pH is dead. Check which years have pH1 data
ph_rows = bottle[bottle['pH1'].notna()].copy()
# Need to join with cast to get Year
cast_slim = cast[['Cst_Cnt', 'Year']].copy()
ph_with_year = ph_rows.merge(cast_slim, on='Cst_Cnt', how='left')
print("pH1 data by year:")
print(ph_with_year['Year'].value_counts().sort_index())
```

```
# pH only exists in 2014-2015. Drop it entirely.
# Now check temporal coverage of remaining variables to inform which to use

# Join bottle with cast to get Year
print("Joining bottle + cast on Cst_Cnt...")
df = bottle[['Cst_Cnt','T_degC','O2ml_L','NO3uM','PO4uM','SiO3uM','ChlorA',
             'T_qual','O_qual','NO3q','PO4q','Depthm']].merge(
    cast[['Cst_Cnt','Sta_ID','Year','Month','Lat_Dec','Lon_Dec']], 
    on='Cst_Cnt', how='inner'
)
print(f"Merged shape: {df.shape}")
print(f"Year range: {df['Year'].min()} – {df['Year'].max()}")
```

```
# Coverage per variable by era: baseline (1949-1990) vs scoring (1991-2021)
baseline = df[df['Year'] <= 1990]
scoring  = df[df['Year'] >  1990]

vars_check = ['T_degC', 'O2ml_L', 'NO3uM', 'PO4uM', 'SiO3uM', 'ChlorA']
print(f"{'Variable':<12} {'Baseline non-null':>20} {'Scoring non-null':>20}")
print("-" * 55)
for v in vars_check:
    b = baseline[v].notna().sum()
    s = scoring[v].notna().sum()
    bp = 100*b/len(baseline)
    sp = 100*s/len(scoring)
    print(f"{v:<12} {b:>10,} ({bp:4.1f}%)   {s:>10,} ({sp:4.1f}%)")
```

```
# Apply quality filtering: drop rows where qual code == 8 or 9
# qual codes are floats; pH dropped; no qual col for SiO3uM/ChlorA
def apply_qual_filter(df):
    d = df.copy()
    # Set measurement to NaN where quality is bad (8=suspect, 9=missing)
    qual_map = {'T_degC': 'T_qual', 'O2ml_L': 'O_qual', 'NO3uM': 'NO3q', 'PO4uM': 'PO4q'}
    for var, qcol in qual_map.items():
        bad = d[qcol].isin([8.0, 9.0])
        d.loc[bad, var] = np.nan
    return d

df_clean = apply_qual_filter(df)

print("After quality filtering:")
for v in ['T_degC', 'O2ml_L', 'NO3uM', 'PO4uM']:
    before = df[v].notna().sum()
    after  = df_clean[v].notna().sum()
    print(f"  {v:<10}: {before:>8,} → {after:>8,}  (dropped {before-after:,})")
```

```
# How many unique stations? What's the depth situation?
# We should probably use surface measurements only (or shallow depth) for consistency
print("Unique stations:", df_clean['Sta_ID'].nunique())
print("\nDepthm distribution (percentiles):")
print(df_clean['Depthm'].describe(percentiles=[.1,.25,.5,.75,.9,.95,.99]))
```

```
# Use near-surface measurements only (0–10m) for consistency across time
# This also matches where ChlorA/nutrients are most ecologically meaningful
df_surface = df_clean[df_clean['Depthm'] <= 10].copy()
print(f"Surface rows (0-10m): {len(df_surface):,}")

baseline_surf = df_surface[df_surface['Year'] <= 1990]
scoring_surf  = df_surface[df_surface['Year'] >  1990]
print(f"  Baseline (≤1990): {len(baseline_surf):,}")
print(f"  Scoring  (>1990): {len(scoring_surf):,}")

print("\nSurface coverage per variable:")
for v in ['T_degC', 'O2ml_L', 'NO3uM', 'PO4uM', 'SiO3uM', 'ChlorA']:
    b = baseline_surf[v].notna().sum()
    s = scoring_surf[v].notna().sum()
    print(f"  {v:<10}: baseline={b:,}  scoring={s:,}")
```

```
# Check station-year coverage in baseline — need 20+ years per station
baseline_station_years = (baseline_surf.groupby('Sta_ID')['Year']
                          .nunique()
                          .rename('baseline_years'))
eligible = baseline_station_years[baseline_station_years >= 20]
print(f"Stations with 20+ baseline years: {len(eligible)} / {len(baseline_station_years)}")
print("\nDistribution of baseline year counts:")
print(baseline_station_years.describe())
```

```
# 106 eligible stations. Check how many also have scoring-period data
scoring_stations = set(scoring_surf['Sta_ID'].unique())
eligible_with_scoring = [s for s in eligible.index if s in scoring_stations]
print(f"Eligible stations that also have scoring data: {len(eligible_with_scoring)}")

# Also check variable coverage for these 106 stations in baseline
baseline_eligible = baseline_surf[baseline_surf['Sta_ID'].isin(eligible.index)]
print("\nVariable coverage for eligible stations (baseline):")
for v in ['T_degC', 'O2ml_L', 'NO3uM', 'PO4uM', 'SiO3uM', 'ChlorA']:
    n = baseline_eligible[v].notna().sum()
    pct = 100 * n / len(baseline_eligible)
    print(f"  {v:<10}: {n:,} ({pct:.1f}%)")
```

```
# Re-filter with 0-50m depth, variables: T_degC, O2ml_L, ChlorA
df_50m = df_clean[df_clean['Depthm'] <= 50].copy()
print(f"0–50m rows: {len(df_50m):,}")

baseline_50 = df_50m[df_50m['Year'] <= 1990]
scoring_50  = df_50m[df_50m['Year'] >  1990]
print(f"  Baseline: {len(baseline_50):,}  |  Scoring: {len(scoring_50):,}")

print("\nVariable coverage (0–50m):")
for v in ['T_degC', 'O2ml_L', 'ChlorA']:
    b = baseline_50[v].notna().sum()
    s = scoring_50[v].notna().sum()
    bp = 100*b/len(baseline_50)
    sp = 100*s/len(scoring_50)
    print(f"  {v:<10}: baseline={b:,} ({bp:.1f}%)  scoring={s:,} ({sp:.1f}%)")
```

```
# Station eligibility: 20+ baseline years at 0-50m
baseline_sta_years = (baseline_50.groupby('Sta_ID')['Year']
                      .nunique()
                      .rename('baseline_years'))
eligible_50 = baseline_sta_years[baseline_sta_years >= 20]
print(f"Eligible stations (20+ baseline years): {len(eligible_50)}")

scoring_stations_50 = set(scoring_50['Sta_ID'].unique())
eligible_with_scoring_50 = [s for s in eligible_50.index if s in scoring_stations_50]
print(f"Eligible stations with scoring data:    {len(eligible_with_scoring_50)}")

# Variable coverage for eligible baseline stations
baseline_elig = baseline_50[baseline_50['Sta_ID'].isin(eligible_50.index)]
print("\nVariable coverage for eligible stations (baseline, 0-50m):")
for v in ['T_degC', 'O2ml_L', 'ChlorA']:
    n = baseline_elig[v].notna().sum()
    pct = 100 * n / len(baseline_elig)
    print(f"  {v:<10}: {n:,} ({pct:.1f}%)")
```

```
# ChlorA baseline is only 23.7% — but that's OK as long as each eligible station
# has SOME baseline ChlorA obs to form a baseline mean/std.
# Check: how many eligible stations have ANY baseline ChlorA?
chlora_baseline_coverage = (baseline_elig.groupby('Sta_ID')['ChlorA']
                             .apply(lambda x: x.notna().sum())
                             .rename('chlora_n'))
print("ChlorA baseline obs per eligible station (distribution):")
print(chlora_baseline_coverage.describe())
print(f"\nStations with 0 baseline ChlorA: {(chlora_baseline_coverage == 0).sum()}")
print(f"Stations with <5 baseline ChlorA: {(chlora_baseline_coverage < 5).sum()}")
print(f"Stations with 5+ baseline ChlorA: {(chlora_baseline_coverage >= 5).sum()}")
```

```
# Great — 105/106 stations have solid ChlorA baseline data.
# Now aggregate per (Sta_ID, Year): mean of each variable across 0-50m bottles
# (multiple bottles per cast, multiple casts per year)

vars_score = ['T_degC', 'O2ml_L', 'ChlorA']
df_50m_eligible = df_50m[df_50m['Sta_ID'].isin(eligible_50.index)]

annual = (df_50m_eligible
          .groupby(['Sta_ID', 'Year', 'Lat_Dec', 'Lon_Dec'])[vars_score]
          .mean()
          .reset_index())

print(f"Annual aggregated rows: {len(annual):,}")
print(f"Year range: {annual['Year'].min()} – {annual['Year'].max()}")
annual.head(5)
```

```
# Lat/Lon can vary slightly per year for same Sta_ID — use median position per station
station_coords = (df_50m_eligible
                  .groupby('Sta_ID')[['Lat_Dec','Lon_Dec']]
                  .median()
                  .reset_index())

# Re-aggregate without lat/lon in groupby keys
annual = (df_50m_eligible
          .groupby(['Sta_ID', 'Year'])[vars_score]
          .mean()
          .reset_index()
          .merge(station_coords, on='Sta_ID'))

print(f"Annual rows: {len(annual):,}  |  Stations: {annual['Sta_ID'].nunique()}")
annual.head()
```

```
# Compute per-station baseline stats (1949–1990)
baseline_annual = annual[annual['Year'] <= 1990]

baseline_stats = (baseline_annual
                  .groupby('Sta_ID')[vars_score]
                  .agg(['mean','std'])
                  .reset_index())
baseline_stats.columns = ['Sta_ID'] + [f'{v}_{s}' for v in vars_score for s in ['mean','std']]

# Drop stations where std == 0 (can't z-score) or NaN baseline
print("Baseline stats shape:", baseline_stats.shape)
print(baseline_stats[['Sta_ID','T_degC_mean','T_degC_std','O2ml_L_mean','O2ml_L_std','ChlorA_mean','ChlorA_std']].describe())
```

```
# Compute health scores for scoring period (1991–2021)
scoring_annual = annual[annual['Year'] > 1990].merge(baseline_stats, on='Sta_ID')

# Z-scores
scoring_annual['z_temp']   = (scoring_annual['T_degC']  - scoring_annual['T_degC_mean'])  / scoring_annual['T_degC_std']
scoring_annual['z_oxygen'] = (scoring_annual['O2ml_L']  - scoring_annual['O2ml_L_mean'])  / scoring_annual['O2ml_L_std']
scoring_annual['z_chlora'] = (scoring_annual['ChlorA']  - scoring_annual['ChlorA_mean'])  / scoring_annual['ChlorA_std']

# Health score: higher temp = worse, lower oxygen = worse, higher chlora = can be bad (algal bloom) or good
# High ChlorA could indicate algal blooms (bad) OR productive ecosystem (neutral/good).
# Using it as: elevated chlora = stress signal (like temp), so: z_chlora subtracts health
# Formula: health = 100 - clip(10*(z_temp - z_oxygen + z_chlora), 0, 100)
# Fill NaN z-scores with 0 (neutral) so missing ChlorA doesn't tank the score
scoring_annual['z_temp_f']   = scoring_annual['z_temp'].fillna(0)
scoring_annual['z_oxygen_f'] = scoring_annual['z_oxygen'].fillna(0)
scoring_annual['z_chlora_f'] = scoring_annual['z_chlora'].fillna(0)

raw_score = 10 * (scoring_annual['z_temp_f'] - scoring_annual['z_oxygen_f'] + scoring_annual['z_chlora_f'])
scoring_annual['health_score'] = (100 - np.clip(raw_score, 0, 100)).round(1)

print(f"Scored rows: {len(scoring_annual):,}")
print(f"Stations with scores: {scoring_annual['Sta_ID'].nunique()}")
print("\nHealth score distribution:")
print(scoring_annual['health_score'].describe())
```

```
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Visualize: mean health score per year across all stations
yearly_mean = scoring_annual.groupby('Year')['health_score'].mean()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Left: mean health score over time
axes[0].plot(yearly_mean.index, yearly_mean.values, marker='o', markersize=3)
axes[0].set_title('Mean Ocean Health Score (1991–2021)')
axes[0].set_ylabel('Health Score (0–100)')
axes[0].set_xlabel('Year')
axes[0].axhline(100, color='green', linestyle='--', alpha=0.4)
axes[0].set_ylim(0, 110)

# Right: distribution of scores
axes[1].hist(scoring_annual['health_score'], bins=30, edgecolor='black')
axes[1].set_title('Distribution of Health Scores')
axes[1].set_xlabel('Health Score')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('health_score_exploration.png', dpi=100)
plt.show()
print("Saved chart.")
```

```
# Check the worst years and stations
print("5 worst years (mean score):")
print(yearly_mean.nsmallest(5))

print("\n10 worst station-years:")
worst = scoring_annual.nsmallest(10, 'health_score')[['Sta_ID','Year','T_degC','O2ml_L','ChlorA','z_temp','z_oxygen','z_chlora','health_score']]
print(worst.to_string())
```

```
# ChlorA is log-normally distributed — standard practice is to work in log space
# Recompute baseline stats with log(ChlorA), then z-score in log space

annual['log_ChlorA'] = np.log(annual['ChlorA'].clip(lower=1e-4))

baseline_annual2 = annual[annual['Year'] <= 1990]
vars_log = ['T_degC', 'O2ml_L', 'log_ChlorA']

baseline_stats2 = (baseline_annual2
                   .groupby('Sta_ID')[vars_log]
                   .agg(['mean','std'])
                   .reset_index())
baseline_stats2.columns = ['Sta_ID'] + [f'{v}_{s}' for v in vars_log for s in ['mean','std']]

# Recompute scoring with log ChlorA
annual['log_ChlorA'] = np.log(annual['ChlorA'].clip(lower=1e-4))
scoring_annual2 = annual[annual['Year'] > 1990].merge(baseline_stats2, on='Sta_ID')

scoring_annual2['z_temp']    = (scoring_annual2['T_degC']      - scoring_annual2['T_degC_mean'])      / scoring_annual2['T_degC_std']
scoring_annual2['z_oxygen']  = (scoring_annual2['O2ml_L']      - scoring_annual2['O2ml_L_mean'])      / scoring_annual2['O2ml_L_std']
scoring_annual2['z_chlora']  = (scoring_annual2['log_ChlorA']  - scoring_annual2['log_ChlorA_mean'])  / scoring_annual2['log_ChlorA_std']

# Fill NaN z-scores with 0 (neutral contribution)
for col in ['z_temp','z_oxygen','z_chlora']:
    scoring_annual2[col] = scoring_annual2[col].fillna(0)

raw2 = 10 * (scoring_annual2['z_temp'] - scoring_annual2['z_oxygen'] + scoring_annual2['z_chlora'])
scoring_annual2['health_score'] = (100 - np.clip(raw2, 0, 100)).round(1)

print("Health score distribution (log ChlorA):")
print(scoring_annual2['health_score'].describe())
print("\n5 worst years:")
print(scoring_annual2.groupby('Year')['health_score'].mean().nsmallest(5))
```

```
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

yearly_mean2 = scoring_annual2.groupby('Year')['health_score'].mean()
axes[0].plot(yearly_mean2.index, yearly_mean2.values, marker='o', markersize=3, color='steelblue')
axes[0].set_title('Mean Ocean Health Score — log(ChlorA) (1991–2021)')
axes[0].set_ylabel('Health Score (0–100)')
axes[0].set_xlabel('Year')
axes[0].set_ylim(0, 110)
axes[0].axhline(100, color='green', linestyle='--', alpha=0.4)

# Annotate known events
axes[0].annotate('1997–98\nEl Niño', xy=(1998, yearly_mean2[1998]), xytext=(2000, 65),
                 arrowprops=dict(arrowstyle='->', color='red'), color='red', fontsize=8)

axes[1].hist(scoring_annual2['health_score'], bins=30, edgecolor='black', color='steelblue')
axes[1].set_title('Distribution of Health Scores (log ChlorA)')
axes[1].set_xlabel('Health Score')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('health_score_v2.png', dpi=100)
plt.show()
print("Saved.")
```

```
import os
os.makedirs('../data/processed', exist_ok=True)

# Save scored output: one row per (station, year)
scores_out = scoring_annual2[['Sta_ID','Year','Lat_Dec','Lon_Dec',
                               'T_degC','O2ml_L','ChlorA',
                               'z_temp','z_oxygen','z_chlora',
                               'health_score']].copy()
scores_out.to_parquet('../data/processed/scores.parquet', index=False)
print(f"Saved scores.parquet: {scores_out.shape}")

# Save baseline stats for reference
baseline_stats2.to_parquet('../data/processed/baseline_stats.parquet', index=False)
print(f"Saved baseline_stats.parquet: {baseline_stats2.shape}")

# Save the full annual aggregated data (includes baseline years for time-series charts)
annual_out = annual[annual['Sta_ID'].isin(eligible_50.index)].merge(station_coords, on='Sta_ID')
annual_out.to_parquet('../data/processed/annual_means.parquet', index=False)
print(f"Saved annual_means.parquet: {annual_out.shape}")
```
